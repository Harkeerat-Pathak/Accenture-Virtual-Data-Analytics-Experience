# Working with Content Dataset. Performing Data Cleaning
1. Dropping the URL and User ID Columns
2. Removing the double quotes " " from Category column
3. Changing the name of column from Type to Content Type

```python
content_clean.head()
```

```python
content_clean = content_clean.drop(columns='URL')
```

```python
content_clean = content_clean.drop(columns='User ID')
```

```python
content_clean.head(55)
```

```python
# Check for missing values
content_clean.isnull().sum()
#There are no null values
```
